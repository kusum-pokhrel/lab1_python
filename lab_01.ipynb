```python
# Basic Data Types of Python
a = 18                  # int : Stores whole numbers
b = 23.99               # float : Stores decimal numbers
c = 5 + 6j              # complex : Stores numbers in a+bj form
s = "python"            # string : Stores text
l = [4, 5, 6]           # list : Ordered and mutable collection
t = (1, 2, 3)           # tuple : Ordered and immutable collection
j = {7, 8, 9}           # set : Unordered collection of unique values
i = True                # bool : Stores True or False
d = {"Name": "kusum", "id": 18}  # dictionary dict : Stores data in key-value pairs
```

```python
# Printing elements of list using for loop
for i in l:
    print(i)
```

```python
# range(start, stop, step) : range() function example
for i in range(0, 10, 2):
    print(i)
```

```python
definition="""
List, dictionary, and set are mutable data types, whereas int, float, string, and tuple are immutable data types.
Mutable data types can be modified after creation and the changes affect the same memory location."""
print(definition)
```

```python
# The type() function is used to find the data type of a variable.
print("The type() function returns the data type of a variable.") 

print(type(a))
print(type(b))
print(type(d))
print(type(l))
print(type(j))
print(type(t))
```

```python
# input() function
print("input() takes input from the user and always returns it as a string")
name = input("Enter your name: ")
print(name)
print(type(name))
print(f"Hello {name}")
```

```python
definition="""
Type casting needed because:
The input() function always takes user input in string format.
To perform mathematical operations, the input must be converted
into int or float using type casting."""
print(definition)
```

```python
print("Type casting with user input")
print(" Without type casting (string concatenation)")
a = input("Enter a number: ")
b = input("Enter a number: ")
print(a + b)

print(" With type casting (addition)")
c = int(input("Enter a number: "))
d = int(input("Enter a number: "))
print(c + d)
```

```python
# Function to calculate Area of Rectangle
def arect(x, y):
    area = x * y
    return area
print(arect(4, 5))
```
